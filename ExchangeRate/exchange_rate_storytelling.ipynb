```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.style as style
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar['rolling_mean']= euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar
```

```python
plt.figure(figsize = (9, 6))
plt.subplot(3, 2, 1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original Values', weight ='bold')

for i, rolling_no in zip([2,3,4,5,6], [7, 30, 50, 100, 365]):
    plt.subplot(3, 2, i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_no).mean())
    plt.title('Rolling Window:'+ str(rolling_no), weight ='bold')
plt.tight_layout()
plt.show()
```

```python
style.use('fivethirtyeight')
```

```python
precovid_2016_2019 = euro_to_dollar.copy(
                    )[(euro_to_dollar['Time'].dt.year >= 2016)
                      & (euro_to_dollar['Time'].dt.year <=2019)]
```

```python
covid_2020 = euro_to_dollar.copy(
                    )[(euro_to_dollar['Time'].dt.year == 2020)]
```

```python
fig, ax= plt.subplots(figsize = (10, 6))
ax.plot(precovid_2016_2019['Time'], precovid_2016_2019['rolling_mean'])
ax.plot(covid_2020['Time'], covid_2020['rolling_mean'])
plt.grid()
plt.show()
```
